```python
from my_lib import *
from IPython.display import clear_output
```

```python
dang_nhap_status = False
user = ''
```

```python
def dang_nhap():
    global dang_nhap_status, user
    print("---GAME LÔ ĐỀ HỌC---")
    print("Vui lòng đăng nhập để chơi game")

    while True:
        username = input("Nhập username: ")
        password = input("Nhập password: ")

        if check_dang_nhap(username, password):
            user = username
            dang_nhap_status = True
            break
        else:
            print("Sai tên người dùng hoặc mật khẩu, vui lòng thử lại.")
    clear_output()
```

```python
def main():
    global dang_nhap_status
    user = ""
    if dang_nhap_status == True:
        dang_nhap()
    if user == 'admin':
        menu_admin()
    else:
        menu()

    if user == 'admin':
        while True:
            lua_chon = int(input("Bạn chọn: "))
            match lua_chon:
            # 1. Tao username
                case 1:
                    clear_output(wait=True)
                    while True:
                        username = input('Username cần tạo: ')
                        if kiem_tra_ton_tai_tai_khoan(username) == True:
                            print('Tài khoản đã tồn tại!')
                        elif username == 'admin':
                            print('Không được tạo tài khoản admin!')
                        else:
                            password = input('Password cần tạo: ')
                            tong_tien = input('Nhập số tiền: ')
                            tao_tai_khoan(username, password, tong_tien)
                        choice = input('Bạn muốn tạo thêm tài khoản không? [không(0); có(1)]: ')
                        if choice == 0:
                            break
                    clear_output(wait=True)
                    main()
                    break
            # 2. Xoa username
                case 2:
                    clear_output(wait=True)
                    while True:
                        username = input('Nhập username cần xóa: ')
                        if kiem_tra_ton_tai_tai_khoan(username) == False:
                            print('Tài khoản bạn muốn xóa không tồn tại!')
                        elif username == 'admin':
                            print('Không được xóa tài khoản admin!')
                        else:
                            xoa_tai_khoan(username)
                        choice = input('Bạn muốn xóa thêm tài khoản không? [không(0); có(1)]: ')
                        if choice == 0:
                            break
                    clear_output(wait=True)
                    main()
                    break
            # 3. Nap tien
                case 3:
                    clear_output(wait=True)
                    while True:
                        username = input('Nhập username cần nạp: ')
                        if kiem_tra_ton_tai_tai_khoan(username) == False:
                            print('Tài khoản bạn muốn nạp tiền không tồn tại!')
                        else:
                            so_tien_nap = input('Nhập số tiền muốn nạp: ')
                            nap_tien(username, so_tien_nap)
                        choice = input('Bạn muốn nạp thêm tiền vào tài khoản không? [không(0); có(1)]: ')
                        if choice == 0:
                            break
                    clear_output(wait=True)
                    main()
                    break
            #4. Thong ke
                case 4:
                    clear_output(wait=True)
                    while True:
                        thong_ke_admin()
                        choice = input('Nhấn Enter để quay lại menu!')
                        if choice == '':
                            break
                    clear_output(wait=True)
                    main()
            #5. Dang xuat
                case 5:
                    dang_nhap_status = False
                    print('Hẹn gặp lại!')
                    break
                    
                case _:
                    print("Nhập sai lựa chọn, vui lòng nhập lại !")
                    
    else:
        while True:
            lua_chon = int(input("Bạn chọn: "))
            match lua_chon:
            # 1. Choi Lo
                case 1:
                    clear_output(wait=True)
                    while True:
                        danh_sach_so_lo = nhap_so_lo()
                        tien_cuoc = nhap_tien_cuoc(user, len(danh_sach_so_lo))
                        game_quay_so(user, danh_sach_so_lo, tien_cuoc)
                        choi_tiep = int(input("Bạn có muốn chơi tiếp không? (0: không, 1: có)"))
                        if choi_tiep == 0:
                            break
                        else:
                            continue
                    clear_output(wait=True)
                    main()
                    break
            #2. Doi Mat Khau
                case 2:
                    clear_output(wait=True)
                    while True:
                        password_cu = input('Nhập password cũ: ')
                        if kiem_tra_tai_khoan(user, password_cu == ''):
                            print('Mật khẩu cũ không đúng!')
                        else:
                            password_moi = input('Nhập password mới: ')
                            doi_mat_khau(user, password_cu, password_moi)
                        choice = input('Bạn muốn đổi mật khẩu nữa không? [không(0); có(1)]: ')
                        if choice == 0:
                            break
                    clear_output(wait=True)
                    main()
                    break    
            #3. Thong Ke                
                case 3:
                    clear_output(wait=True)
                    while True:
                        thong_ke_user(user)
                        choice = input('Nhấn Enter để quay lại menu!')
                        if choice == '':
                            break
                    clear_output(wait=True)
                    main()
                    break
            #4. Dang Xuat
                case 4:
                    dang_nhap_status = False
                    print('Hẹn gặp lại!')
                    break
                    
                case _:
                    print("Nhập sai lựa chọn, vui lòng nhập lại !")
```
