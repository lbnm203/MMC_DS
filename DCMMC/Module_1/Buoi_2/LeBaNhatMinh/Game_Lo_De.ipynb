```python
import random
from IPython.display import clear_output

tong_tien = 1000000

global tong_tien
def quay_so():
    giai_nhat = random.randint(10,99)
    giai_hai = random.randint(10,99)
    giai_ba = random.randint(10,99)
    giai_tu = random.randint(10,99)
    giai_nam = random.randint(10,99)
    giai_sau = random.randint(10,99)
    giai_bay = random.randint(10,99)
    return giai_nhat, giai_hai, giai_ba, giai_tu, giai_nam, giai_sau, giai_bay

def in_ds_giai(ds_giai):
    giai = 1
    for so in ds_giai:
        print(f"Giai {giai}: {so}")
        giai += 1

def nhap_lo():
    while True:
        so_lo = input("Nhp s l mun cc (10-99): ")
        if so_lo.isdigit() and 10 <= int(so_lo) <= 99:
            return int(so_lo)
        else:
            print("Nhp sai nh dng, nhp li!")

def nhap_tien_cuoc():
    
    while True:
        tien_cuoc = input("Nhp tin mun cc (tin cc < tng tin bn c): ")
        if tien_cuoc.isdigit() and int(tien_cuoc) <= tong_tien:
            return int(tien_cuoc)
        else:
            print("Nhp sai nh dng hoc s tin cn li < tng tin. Nhp li")

def man_hinh_quay_so(so_lo, tien_cuoc):
    global tong_tien
    ds_giai = quay_so()
    dem_so_lo = 0
    print(f"S l bn chi l: {so_lo}")
    in_ds_giai(ds_giai)
    for giai in ds_giai:
        if so_lo == giai:
            dem_so_lo += 1
    if dem_so_lo > 0:
        tong_tien += (tien_cuoc * dem_so_lo) * 70
        print(f"Bn  trng {dem_so_lo} s!")
        print(f"S tin bn trng l {(tien_cuoc * dem_so_lo) * 70}")
        print(f"Tng tin bn c l {tong_tien}")
    else:
        print("Bn  thua l!")
        tong_tien -= tien_cuoc
        print(f"Tng tin bn cn l {tong_tien}")

def nap_tien():
    global tong_tien
    while True:
        so_tien_nap = int(input("Nhp s tin m bn mun np: "))
        tong_tien += so_tien_nap
        print(f"Tng tin sau khi bn np l {tong_tien}")
        nap_tiep = int(input("Bn c mun np tip khng? (0: khng, 1: c)"))
        if nap_tiep == 0:
            break
        else:
            continue

def menu():
    print("GAME L  ")
    print("1. Chi l")
    print("2. Np tin")
    print("3. Thot game")

def main():
    menu()
    while True:
        lua_chon = int(input("Bn chn: "))
        match lua_chon:
            case 1:
                clear_output(wait=True)
                while True:
                    tien_cuoc = nhap_tien_cuoc()
                    so_lo = nhap_lo()
                    man_hinh_quay_so(so_lo, tien_cuoc)
                    choi_tiep = int(input("Bn c mun chi tip khng? (0: khng, 1: c)"))
                    if choi_tiep == 0:
                        break
                clear_output(wait=True)
                main()
                break
            case 2:
                clear_output(wait=True)
                nap_tien()
                clear_output(wait=True)
                main()
                break
            case 3:
                print("Cho bn, hn gp li!")
                break
            case _:
                print("Nhp sai la chn, vui lng nhp li !")

main()
```
